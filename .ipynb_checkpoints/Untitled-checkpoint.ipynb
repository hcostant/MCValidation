```python

def make_random_ipol_table(x, y, n):
#     print(x)
#     print(y)
#     print(n)
#    double *x2, *y2;
#    double xlow, xhigh;
#    int i;
   
#    if ( n<2 )
#       return -1;
    xlow = x[0] - 0.5 * (x[1]-x[0])
    xhigh = x[n-1] + 0.5 * (x[n-1]-x[n-2])
#     print('xlow =%s, xhigh=%s' % (xlow, xhigh))
    x2 = y2 = np.zeros(n+1)
#    if ( (x2=(double *)calloc((size_t)(n+1),sizeof(double))) == NULL ||
#         (y2=(double *)calloc((size_t)(n+1),sizeof(double))) == NULL )
#    {
#       if ( x2 != NULL )
#          free(x2);
#       if ( y2 != NULL )
#          free(y2);
#       return -1;
#    }
   
    y2[0] = 0.
    y2[1] = y[0]
#    /* This 'integration' now tries to account for cases where */
#    /* the step size change. That can only be an approximation. */
#    for (i=1; i<n; i++)
    for i in range(0,n):
        if i+1<n:
            y2[i+1] = y2[i] + y[i] * 0.5*((x[i+1])-x[i-1])/(x[1]-x[0])
#             print('i+1<n: y2= %s' % y2[i+1])
        else:
            y2[i+1] = y2[i] + y[i] * 0.5*((xhigh)-x[i-1])/(x[1]-x[0])
#             print('i+1>n: y2= %s' % y2[i+1])

#    for (i=0; i<n; i++)
#       y2[i] /= y2[n];
    xr=[]
    yr=[]
    for i in range(0,n):
#         print('x=%s y2i(%s)/y2n(%s)= %s' % (0.5*(x[i-1]+x[i]), y2[i],y2[n],y2[i]/y2[n]))
        xr.append(0.5*(x[i-1]+x[i]))
        yr.append(y2[i]/y2[n])
#         y2[i] = y2[i]/ y2[n]
        
    
#     y2[n] = 1.

    xr[0] = xlow;
#    for (i=1; i<n; i++)
#     for i in range(0,n):
#         x2[i] = 0.5*(x[i-1]+x[i])
    xr[-1] = xhigh
#     xr = x2;
#     yr = y2;
#     print('XR', xr) 
#     print('YR', yr)
    return xr, yr
```

```python
# /* ------------------------- random_from_ipol_table ---------------------- */
# /**
#  *  @short Fast random number table lookup.
#  *
#  *  Get a random number from a lookup table as created by
#  *  make_random_ipol_table() for a tabulated probability distribution.
#  *
#  *  @param  xr   Table as returned from make_random_ipol_table()
#  *  @param  yr   Table as returned from make_random_ipol_table()
#  *  @param  n	 as provided to make_random_ipol_table()
#  *  @param  randnr  A uniform pseudorandom number in [0:1]
#  *  @return random number according to table
#  *
# */

def random_from_ipol_table(xr, yr,randnr):
    fun = interp1d(yr,xr);
    return fun(randnr)
```
